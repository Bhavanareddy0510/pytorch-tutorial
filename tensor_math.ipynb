```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#addition
tensor_a + tensor_b
```

```python
#subtraction
tensor_a - tensor_b
```

```python
#addition longhand
torch.add(tensor_a , tensor_b)
```

```python
tensor_b - tensor_a
```

```python
torch.sub(tensor_b , tensor_a)
```

```python
tensor_a * tensor_b
```

```python
torch.mul(tensor_a , tensor_b)
```

```python
#divison
tensor_b / tensor_a
```

```python
#divison longhand
torch.div(tensor_b , tensor_a)
```

```python
#remmainder modules
tensor_b % tensor_a
```

```python
#remainder longhand
torch.remainder(tensor_b , tensor_a)
```

```python
#exponents/power
torch.pow(tensor_a , tensor_b)
```
